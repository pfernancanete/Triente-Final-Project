### Import libraries

```python
import sqlalchemy as alch
from getpass import getpass
import pandas as pd
```

### Nos conectamos al servidor
Para conectarnos a una base de datos de SQL con Python utilizamos SQL Alchemy

```python
password = getpass("Introduce tu pass de sql: ")
dbName="triente"
connectionData=f"mysql+pymysql://root:{password}@localhost/{dbName}"
```

Lo primero que hay que hacer para trabajar con SQLAlchemy es crear un engine. El engine es el punto de entrada a la base de datos, es decir, el que permite a SQLAlchemy comunicarse con esta. El motor se usa principalmente para manejar dos elementos: los pools de conexiones y el dialecto a utilizar.

```python
engine = alch.create_engine(connectionData)
print("me conecté")
```

```python
# Creamos la base de datos triente
engine.execute("""
    CREATE DATABASE triente ;

""")
```

```python
def view_all_comps(razon_social):
    company = engine.execute(f"""
    SELECT razon_social, cif, domicilio, municipio, codigo_postal, provincia, pais, telefono, web, email, cnae, descripcion_actividad 
    FROM Company
    WHERE razon_social = "{razon_social}"
    """)
    try:
        data = company.fetchall()
        return data
    except:
        print('Ese usuario no está registrado')
```

```python
view_all_comps('ACIEROID, S.A.')
```

```python
def get_comp_by_country(pais):
    pais = engine.execute(f"""
    SELECT razon_social 
    FROM Company 
    WHERE pais = "{pais}"
    """)
    try:
        data = pais.fetchall()
        return data
    except:
        print('No existen empresas en ese pais')
```

```python
get_comp_by_country('España')
```

```python
def get_comp_by_provincia(provincia):
    provincia = engine.execute(f"""
    SELECT razon_social 
    FROM Company 
    WHERE provincia = "{provincia}"
    """)
    try:
        data = provincia.fetchall()
        return data
    except:
        print('No existen empresas en esa provincia')
```

```python
get_comp_by_provincia('madrid')
```

```python
def get_comp_by_actividad(actividad):
    actividad = engine.execute(f"""
    SELECT razon_social 
    FROM Company 
    WHERE actividad_1 OR actividad_2 OR actividad_3 OR actividad_4 OR actividad_5 OR actividad_6 = "{actividad}"
    """)
    try:
        data = actividad.fetchall()
        return data
    except:
        print('No existen empresas que se dediquen a esa actividad')
```

```python
get_comp_by_actividad('Arquitectura')
```
